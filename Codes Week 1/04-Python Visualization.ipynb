# Data Visualization with Python: A Hands-On Tutorial


This notebook is designed to help you understand the basics of data visualization using Python. 
We will cover some of the most popular libraries: pandas, matplotlib, and seaborn. 
By the end of this tutorial, you will have a good understanding of how to create and customize different types of plots.


```python

# Let's start by importing the necessary libraries.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```


## Section 1: Introduction to Pandas DataFrames

Before we dive into plotting, let's quickly review how to work with data in pandas.
We'll create a simple DataFrame to work with.


```python

data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
}

df = pd.DataFrame(data)
print("DataFrame Example:")
print(df)
```


Pandas DataFrame is the key data structure used for data manipulation. 
Now, let's move on to plotting.



## Section 2: Basic Plotting with Matplotlib

Matplotlib is the most fundamental plotting library in Python.
It provides a lot of flexibility but also requires more code for customization.

Let's start with a simple bar chart.


```python

plt.figure(figsize=(8, 5))
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.title('Simple Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```


**Explanation:**
- `plt.bar()` is used to create a bar chart.
- The `color` parameter allows us to change the color of the bars.
- We use `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` to add a title and labels.



## Section 3: Customizing Matplotlib Plots

Customization is key to making your plots informative and visually appealing.

Let's add more customization to our bar chart.


```python

plt.figure(figsize=(8, 5))
plt.bar(df['Category'], df['Values'], color='orange', edgecolor='black', linewidth=1.5)
plt.title('Customized Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```


**Explanation:**
- `edgecolor` and `linewidth` parameters are used to customize the edges of the bars.
- `plt.grid()` adds a grid to the plot, which can improve readability.



## Section 4: Line Plot with Matplotlib

Line plots are useful for showing trends over time.

Let's create a simple line plot.


```python

plt.figure(figsize=(8, 5))
plt.plot(df['Category'], df['Values'], marker='o', color='green')
plt.title('Simple Line Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```


**Explanation:**
- `plt.plot()` is used to create a line plot.
- The `marker` parameter adds markers at each data point.



## Section 5: Introduction to Seaborn

Seaborn is a high-level data visualization library based on Matplotlib.
It provides a more user-friendly interface and beautiful default styles.

Let's use seaborn to create a bar plot.


```python

sns.set(style="whitegrid")  # Set style
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Values', data=df, palette='Blues_d')
plt.title('Seaborn Bar Plot')
plt.show()
```


**Explanation:**
- `sns.barplot()` creates a bar plot with much simpler code.
- `palette` parameter allows us to set a color palette.



## Section 6: Advanced Plotting with Seaborn

Seaborn makes it easy to create complex and informative visualizations.

Let's create a box plot using an example dataset.


```python

# Loading an example dataset
iris = sns.load_dataset("iris")

plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='sepal_length', data=iris, palette='Set3')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```


**Explanation:**
- `sns.boxplot()` is used to create a box plot.
- Box plots are great for visualizing the distribution of data and identifying outliers.



## Section 7: Pair Plot for Multivariate Analysis

Pair plots are used to visualize pairwise relationships in a dataset.

Let's create a pair plot for the Iris dataset.


```python

plt.figure(figsize=(8, 5))
sns.pairplot(iris, hue='species', palette='husl')
plt.show()
```


**Explanation:**
- `sns.pairplot()` creates a matrix of plots for each pair of variables.
- The `hue` parameter allows us to color the plots by species.



## Section 8: Heatmaps for Correlation Matrices

Heatmaps are great for visualizing matrices, such as correlation matrices.

Let's create a heatmap of the correlation matrix for the Iris dataset.


```python

plt.figure(figsize=(8, 5))
corr_matrix = iris.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Iris Correlation Matrix')
plt.show()
```


**Explanation:**
- `sns.heatmap()` is used to create a heatmap.
- The `annot=True` parameter displays the correlation values on the heatmap.



## Section 9: Conclusion

We've seen how to create and customize various types of plots, from simple bar charts to more advanced visualizations like pair plots and heatmaps.
With these tools, you can create informative and visually appealing data visualizations to better understand and present your data.
